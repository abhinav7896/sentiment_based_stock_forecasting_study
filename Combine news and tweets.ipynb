```python
import pandas as pd

def timestamp(date):
    return pd.Timestamp(date)

fintweets = pd.read_json('fintweets.json')
data1 = fintweets.to_dict()
data1['microsoft']
news = pd.read_json('nytimes.json')
data2 = news.to_dict()
```

```python
#Combine tweets and news of respective companies for the respective dates
for company in data1:
    for date2 in data2[company]:
        if(date2 in data1[company]):
            datatype1 = type(data1[company][date2])
            datatype2 = type(data2[company][date2])                 
            if(datatype1 != list or datatype2 != list):
                data1[company][date2] = data2[company][date2]
            else:
                data1[company][date2].extend(data2[company][date2])
        else:
            data1[company][date2] = data2[company][date2].copy()
data1
```

```python
tweetsnnews = pd.DataFrame.from_dict(data1)
tweetsnnews.to_json('tweetsnnews.json')
```
