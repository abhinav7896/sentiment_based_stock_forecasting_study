```
# File location and type
file_location = "/FileStore/tables/tweets_clean2.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","
```

```
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
df.count()
```

```
df.createOrReplaceTempView('tweets')
res = spark.sql('select * from tweets where _c1=0')
display(res)
```

```
from pyspark.ml.feature import Tokenizer, RegexTokenizer
from pyspark.sql.functions import col, udf
from pyspark.sql.types import IntegerType
from pyspark.ml.feature import Word2Vec
from pyspark.sql.functions import col, udf, lit
import numpy as np
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler

vocab_size=100000
emdedding_size=128
tokenizer = Tokenizer(inputCol="_c2", outputCol="Tokenized texts")

regexTokenizer = RegexTokenizer(inputCol="_c2", outputCol="Tokenized texts", pattern="\\W")

df = df.dropna()
tokenized = tokenizer.transform(df)
tokenized = tokenized.select("Tokenized texts", "_c1").dropna()
word2Vec = Word2Vec(vectorSize=emdedding_size, minCount=0, inputCol="Tokenized texts", outputCol="embeddings")
word2VecModel = word2Vec.fit(tokenized)

result = word2VecModel.transform(tokenized)
assembler = VectorAssembler(
    inputCols=["embeddings"],
    outputCol="X")
data = assembler.transform(result).select('X', '_c1').withColumnRenamed('_c1', 'Y')
data.show()
data.count()
```

```
data = data.select(data.X, data.Y.cast('float'))
data.printSchema()
```

```
train, test = data.randomSplit(weights = [0.80, 0.20], seed = 13)
```

```
from tensorflow import keras
import tensorflow as tf
import horovod.spark.keras as hvd
from horovod.spark.common.store import DBFSLocalStore

emdedding_size=128
inputs = keras.layers.Input(shape=(1,emdedding_size))
h=keras.layers.LSTM(emdedding_size, dropout=0.8, recurrent_dropout=0.8)(inputs)
h=keras.layers.Dense(64, activation='relu')(h)
outputs= keras.layers.Dense(1, activation='sigmoid')(h)
model = keras.models.Model(inputs, outputs)

optimizer = keras.optimizers.Adam(lr=0.01)
loss = 'binary_crossentropy'
dbfs_path = '/dbfs/train'
store = DBFSLocalStore(dbfs_path)
keras_estimator = hvd.KerasEstimator(
    num_proc=2,
    store=store,
    model=model,
    optimizer=optimizer,
    loss=loss,
    feature_cols=['X'],
    label_cols=['Y'],
    batch_size=5000,
    epochs=10)


keras_model = keras_estimator.fit(train)
predict_df = keras_model.transform(test)
```

```
predict_df.show()
```

```
predict_df.filter("Y = 0").show()
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
predict_df = predict_df.select(predict_df.X, predict_df.Y, predict_df.Y__output.cast('double'))
predict_df.printSchema()
evaluator = BinaryClassificationEvaluator(rawPredictionCol='Y__output', labelCol='Y', metricName='areaUnderROC')

print('Test accuracy:', evaluator.evaluate(predict_df))
```

```
# Predict sentiment for given text

from pyspark.ml.feature import Word2Vec
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import Tokenizer, RegexTokenizer
from pyspark.sql.types import StringType
from pyspark.sql.functions import col, avg

def getSentimentScore(info):
  emdedding_size=128
  if(info == []):
    return -1    #Dates without news or tweet data. Will be replaced with average score in he end
  columns = ["text"]
  documentDF = spark.createDataFrame(info, StringType()).withColumnRenamed('value', 'text').dropna()
  tokenizer = Tokenizer(inputCol="text", outputCol="tokens")

  documentDF = documentDF.dropna()
  tokenized = tokenizer.transform(documentDF)
  tokenized = tokenized.select("text", "tokens").dropna()

  w2v = Word2Vec(vectorSize=emdedding_size, minCount=0, inputCol="tokens", outputCol="embeddings")
  m = w2v.fit(tokenized)

  x = m.transform(tokenized)
  assembler = VectorAssembler(
      inputCols=["embeddings"],
      outputCol="X")
  x = assembler.transform(x)
  pred = keras_model.transform(x)
  score = pred.select(avg("Y__output")).collect()[0][0]
  return score
```
