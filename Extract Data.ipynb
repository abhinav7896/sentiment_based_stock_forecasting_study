```python
#News from New York Times
import requests
import time

FROM_DATE = "20191207"
TILL_DATE = "20201207"
keywords = {'microsoft':['microsoft', 'MSFT', 'Nadella', 'Xbox', 'Teams', 'Windows OS'], 
            'tesla': ['tesla', 'TSLA', 'Elon', 'Musk'], 
            'google': ['google', 'DeepMind', 'Pichai'],
            'apple': ['apple', 'Tim Cook', 'iphone12', 'iphone', 'ipad', 'macbook']
           }
news = dict()
for keyword in keywords:
    print('For keyword: ', keyword)
    news[keyword] = {}
    for pno in range(30):
        if(pno%10 == 9):
            time.sleep(120)                        
        newsNYT = requests.get('https://api.nytimes.com/svc/search/v2/articlesearch.json?q={}&begin_date={}&end_date={}&sort=relevance&page={}&api-key=c28QQVh56M7ePFGo5ay0WrRZGAOtjTqm '.format(keyword, FROM_DATE, TILL_DATE, pno))
        print('HTTP status: ', newsNYT.status_code)
        for doc in newsNYT.json()['response']['docs']:
            date = doc['pub_date'].split('T')[0]
            if(news[keyword].get(date) == None):
                news[keyword][date] = [doc['lead_paragraph']]
            else:
                news[keyword][date].append(doc['lead_paragraph'])
              
# Removes non ascii characters while writing into the file
with open('news_nytimes.json', 'w', encoding='ascii', errors='ignore') as file:
    json.dump(news, file)        
news
```

```python
import twint
import nest_asyncio
import json
nest_asyncio.apply()
```

```python
fin_news = ['Stocktwits', 'Benzinga', 'CNBC', 'BreakoutStocks', 'bespokeinvest', 'WSJmarkets', 'Stephanie_Link', 'nytimesbusiness', 
           'IBDinvestors', 'WSJDealJournal']

keywords = {'microsoft':['microsoft', 'MSFT', 'Nadella', 'Xbox', 'Teams', 'Windows OS'], 
            'tesla': ['tesla', 'TSLA', 'Elon', 'Musk'], 
            'google': ['google', 'DeepMind', 'Pichai'],
            'apple': ['apple', 'Tim Cook', 'iphone12', 'iphone', 'ipad', 'macbook']
           }
info = {}
config = twint.Config()
for keyword in keywords:
    info[keyword] = {}
    for channel in fin_news:
        tweetsdata = []
        config.Username = channel
        config.Search = keyword
        config.Lang = "en"
        config.Limit = 10000
        config.Min_likes = 0 
        config.Lower = True
        config.Store_object = True
        config.Store_object_tweets_list = tweetsdata
        twint.run.Search(config)
        for item in tweetsdata:
            date = item.datestamp
            if(info[keyword].get(date) == None):
                info[keyword][date] = [item.tweet]
            else:
                info[keyword][date].append(item.tweet)
                
with open('fintweets.json', 'w', encoding='ascii', errors='ignore') as file:
    json.dump(info, file)
```
