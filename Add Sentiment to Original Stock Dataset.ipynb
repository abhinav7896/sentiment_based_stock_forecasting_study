```python
import pandas as pd
import warnings
from pandas.core.common import SettingWithCopyWarning

warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

df = pd.read_csv('stocks.csv')
tickers = ['MSFT', 'TSLA', 'GOOGL', 'AAPL']
stockdata = {}
for ticker in tickers:
    stockdata[ticker] = df.loc[df['Stock'] == ticker]
    stocksent = pd.read_csv(ticker+'_scores.csv')
    stocksent = stocksent[['Stock', 'Date', 'Sentiment Score']]
    stocksent = stocksent.set_index(stockdata[ticker].index)
    stockdata[ticker]['Sentiment Score'] = stocksent['Sentiment Score']
    stockdata[ticker].to_csv(ticker+'_with_sentiment.csv')
stockdata['AAPL']
```
